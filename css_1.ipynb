```python
# chirp spread spectrum modulation
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

plt.close("all")

# time signal with 100 hz sampling rate
dt1 = 0.01
fs1 = 1/dt1
n1 = np.arange(0, 10, dt1)
len1 = len(n1)
print("sampling freq:", fs1)

# chirp signals
# +chirp signal for ones
x1 = sp.signal.chirp(n1, 0, 10, 10)
# -chirp signal for zeros
x2 = sp.signal.chirp(n1, 10, 10, 0)

# plot of chirp signals
plt.figure()
plt.subplot(2,1,1)
plt.plot(n1, x1)
plt.title("plot of +chirp signal")
plt.xlabel("time")
plt.ylabel("amplitude")

plt.subplot(2,1,2)
plt.plot(n1,x2)
plt.title("plot of -chirp signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.tight_layout()
plt.show()
```

```python
# spectrogram of chirp signals
plt.figure()
plt.subplot(2,1,1)
plt.specgram(x1, Fs=fs1, NFFT=128, noverlap=64)
plt.title("spectrogram of +chirp signal")
plt.xlabel("time")
plt.ylabel("frequency")

plt.subplot(2,1,2)
plt.specgram(x2, Fs=fs1, NFFT=128, noverlap=64)
plt.title("spectrogram of -chirp signal")
plt.xlabel("time")
plt.ylabel("frequency")
plt.tight_layout()
plt.show()
```

```python
# magnitude spectrum of chirp signals
plt.figure()
plt.subplot(2,1,1)
plt.magnitude_spectrum(x1, Fs=fs1)

plt.subplot(2,1,2)
plt.magnitude_spectrum(x1, Fs=fs1)
plt.tight_layout()
plt.show()
```

```python
# message signal
message1 = np.random.randint(2, size=10)
message1 = np.array(message1)
print(message1)

plt.figure()
plt.step(range(len(message1)), message1)
plt.title("message signal")
plt.xlabel("sample number")
plt.ylabel("amplitude")
plt.show()
```

```python
# initialize modulated signal
modulated1 = []
modulated1 = np.array(modulated1)
n2 = np.arange(0, 100, dt1)

# generating modulating signal via message signal
for i1 in message1:
    if i1 == 1:
        modulated1 = np.append(modulated1, x1)
    elif i1 == 0:
        modulated1 = np.append(modulated1, x2)

# plot of modulated signal
plt.figure()
plt.subplot(2,1,1)
plt.plot(n2, modulated1)
plt.title("plot of modulated signal")
plt.xlabel("time")
plt.ylabel("amplitude")

# magnitude spectrum of modulated signal
plt.subplot(2,1,2)
plt.magnitude_spectrum(modulated1, Fs=fs1)
plt.tight_layout()
plt.show()
```

```python
# spectorgram of modulated signal
plt.figure()
plt.specgram(modulated1, Fs=fs1, NFFT=128, noverlap=64)
plt.title("spectrogram of modulated signal")
plt.xlabel("time")
plt.ylabel("frequency")
plt.show()
```
