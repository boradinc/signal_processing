```python
# frequency hopping spread spectrum modulation with frequency shift keying modulation
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

plt.close("all")

# time signal with 100hz sampling rate
dt1 = 0.01
fs1 = 1/dt1
n1 = np.arange(0, 2*np.pi, dt1)

# message signal
message1 = np.random.randint(2, size=10)
print("message signal:", message1)

plt.figure()
plt.step(range(len(message1)), message1)
plt.title("message signal")
plt.xlabel("sample number")
plt.ylabel("amplitude")
plt.show()
```

```python
# creating 5 channels
# each channel has 2 additional channels for 2 bits
ch0_1 = np.cos(2 *2*np.pi*n1)
ch0_2 = np.cos(4 *2*np.pi*n1)
ch1_1 = np.cos(6 *2*np.pi*n1)
ch1_2 = np.cos(8 *2*np.pi*n1)
ch2_1 = np.cos(10*2*np.pi*n1)
ch2_2 = np.cos(12*2*np.pi*n1)
ch3_1 = np.cos(14*2*np.pi*n1)
ch3_2 = np.cos(16*2*np.pi*n1)
ch4_1 = np.cos(18*2*np.pi*n1)
ch4_2 = np.cos(20*2*np.pi*n1)

# channel selection for fhss modulation
channel1 = np.random.randint(5, size=5)
channel1 = np.append(channel1, channel1)
print("channel selection:", channel1)

# initialize modulated signal
modulated1 = []
modulated1 = np.array(modulated1)

for i1 in range(len(channel1)):
    if channel1[i1] == 0:
        if message1[i1] == 0:
            modulated1 = np.append(modulated1, ch0_1)
        else:
            modulated1 = np.append(modulated1, ch0_2)
    elif channel1[i1] == 1:
        if message1[i1] == 0:
            modulated1 = np.append(modulated1, ch1_1)
        else:
            modulated1 = np.append(modulated1, ch1_2)
    elif channel1[i1] == 2:
        if message1[i1] == 0:
            modulated1 = np.append(modulated1, ch2_1)
        else:
            modulated1 = np.append(modulated1, ch2_2)
    elif channel1[i1] == 3:
        if message1[i1] == 0:
            modulated1 = np.append(modulated1, ch3_1)
        else:
            modulated1 = np.append(modulated1, ch3_2)
    elif channel1[i1] == 4:
        if message1[i1] == 0:
            modulated1 = np.append(modulated1, ch4_1)
        else:
            modulated1 = np.append(modulated1, ch4_2)

# time signal for modulated signal
n2 = np.array([])

for i1 in range(10):
    a1 = n1 + ((i1+1)*2*np.pi)
    n2 = np.append(n2, a1)

plt.figure()
plt.plot(n2, modulated1)
plt.title("modulated signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

```python
plt.figure()
plt.specgram(modulated1, Fs=fs1)
plt.title("spectrogram of modulated signal")
plt.xlabel("time")
plt.ylabel("frequency")
plt.show()
```
